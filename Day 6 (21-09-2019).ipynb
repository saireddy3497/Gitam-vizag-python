### Visualizing Data
- An intersting way to display Dataframe
- with matplotlib

```python
import matplotlib.pyplot as plt
a1 = [1,2,3,4]
a2 = [5,6,7,8]
plt.figure(figsize=(5,5))
plt.plot(a1,a2,color='green')
plt.xlabel('x-Axis')
plt.ylabel('y-Axis')
plt.show()
```

```python
divisions = ['A','B','C','D']
marks = [79,82,75,63]
avg_marks = [56,72,78,89]
index = np.arange(len(divisions))
plt.bar(divisions,marks,color="#1412cd",width = 0.380)
```

```python
plt.barh(divisions,marks,color='pink')
plt.show()
```

```python
language = ['python','java','c','c++']
rating =[ 45,70,75,80]
plt.pie(rating,labels = language,shadow=True,startangle=90,autopct='%1.1f%%')
plt.legend(loc='best')
plt.show()
```
